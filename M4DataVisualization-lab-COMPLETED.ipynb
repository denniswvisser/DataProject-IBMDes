### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
conn = sqlite3.connect("m4_survey_data.sqlite")
# Read the entire "Master" table into a DataFrame
df = pd.read_sql_query("SELECT * FROM Master", conn)

# Close the database connection
conn.close()

# List of database names to search for
database_names = ["PostgreSQL", "MongoDB", "MySQL", "Microsoft SQL Server", "Redis", 
                  "Elasticsearch", "Cassandra", "Neo4j", "SQLite", "Oracle Database"]

# Create a new DataFrame to store the matching responses
matching_responses = pd.DataFrame(columns=df.columns)

# Search for responses containing the database names in all columns
for db_name in database_names:
    matching_df = df[df.apply(lambda row: db_name in ' '.join(map(str, row)), axis=1)]
    matching_responses = pd.concat([matching_responses, matching_df])

# Display the matching responses
print(matching_responses)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
DATABASE_URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the "Master" table into a DataFrame
df = pd.read_sql_query("SELECT Age, WorkWeekHrs, CodeRevHrs FROM Master", conn)

# Close the database connection
conn.close()

# Filter the data for the age group 30 to 35
age_group_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median of WorkWeekHrs and CodeRevHrs for the age group
median_data = age_group_df.groupby('Age').median()

# Plot the stacked chart
median_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age (years)')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend(["WorkWeekHrs", "CodeRevHrs"])
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
DATABASE_URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the "Master" table into a DataFrame
df = pd.read_sql_query("SELECT Age, ConvertedComp FROM Master", conn)

# Close the database connection
conn.close()

# Filter the data for ages from 45 to 60
age_group_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_data = age_group_df.groupby('Age').median()

# Plot the line chart
median_data.plot(kind='line', marker='o', figsize=(10, 6))
plt.xlabel('Age (years)')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()








```
