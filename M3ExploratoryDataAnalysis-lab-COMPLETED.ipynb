## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
plt.figure(figsize=(8, 6))
plt.boxplot(df['ConvertedComp'], vert=True)
plt.xlabel('Converted Compensation (USD)')
plt.title('Box Plot of Converted Compensation')
plt.show()





```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'], vert=False)
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

NOTES:
    #Use plt.boxplot() to create the box plot for our 'ConvertedComp' column. 
#The vert=False parameter makes the plot horizontal for better visibility 
#In the resulting box plot, the box represents the interquartile range (IQR) 
#which spans from the first quartile (Q1) to the third quartile (Q3). 
#The horizontal line inside the box represents the median (Q2).
#Outliers are plotted as individual points beyond the whiskers of the box. 
#The whiskers are usually located at 1.5 times the IQR from the first and third quartiles. 
#Any point beyond the whiskers is considered an outlier and marked as such.
#If you see any points beyond the whiskers in the box plot, those are potential outliers 
#in the 'ConvertedComp' column. Outliers could be extreme high or low values compared 
#to the rest of the data points. 
#Keep in mind that outliers may or may not represent errors or anomalies in the data 
#and may require further investigation.



```python
summary = df['ConvertedComp'].describe()

# Round the results to two decimals
summary_rounded = summary.round(2)

# Display the summary
print(summary_rounded)
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
q1 = df['ConvertedComp'].quantile(0.25)

# Calculate the 75th percentile (Q3)
q3 = df['ConvertedComp'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# Display the IQR
print("Interquartile Range (IQR) for 'ConvertedComp':", iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr

# Display the bounds
print("Upper Bound:", upper_bound)
print("Lower Bound:", lower_bound)



```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]

# Count the number of outliers
num_outliers = len(outliers)

# Display the number of outliers
print("Number of outliers in 'ConvertedComp':", num_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
print(df_no_outliers.head())
```

```python
# your code goes here
transposed_df1 = df_no_outliers.transpose()

# Using the .T attribute
transposed_df2 = df_no_outliers.T

# Display the transposed DataFrames
print("Transposed DataFrame using transpose() method:")
print(transposed_df1)

print("\nTransposed DataFrame using .T attribute:")
print(transposed_df2)
##Both transpose() and .T will produce the same result, which is the DataFrame with rows and columns switched. The resulting DataFrame, whether stored in transposed_df1 or transposed_df2, will have the rows of the original DataFrame 'df_no_outliers' as columns and vice versa. The original DataFrame will remain unchanged.





```
